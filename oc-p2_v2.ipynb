```python
df_nonan_dropped_columns_grade_a_only = df_nonan_dropped_columns[df_nonan_dropped_columns['nutrition_grade_fr'] == 'a']
```

```python
df_nonan_dropped_columns_grade_a_only.head()
```

```python
# create a confusion matrix
from yellowbrick.features import Rank2D

r2D = Rank2D(algorithm='pearson')
r2D.fit(df_nonan_dropped_columns_grade_a_only)
r2D.transform(df_nonan_dropped_columns_grade_a_only)
r2D.show()
```
